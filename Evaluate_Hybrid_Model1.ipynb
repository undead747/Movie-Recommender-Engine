```python
import pandas as pd
import numpy as np
import pickle
```

```python
f = open('KNN_Mean_Pred_Rating_Test.pckl', 'rb')
KNN_predict_rating_result = pickle.load(f)
f.close()
```

```python
len(KNN_predict_rating_result)
```

```python
test = pd.read_csv('../the-movies-dataset/ratings_test.csv')
```

```python
test['KNN-pred'] = KNN_predict_rating_result.values()    
test
```

```python
se = 0
for index, row in test.iterrows():
    e = row.rating - row['KNN-pred']
    se += e*e
np.sqrt(se/test.shape[0])    
```

```python
f = open('tf_idf_pred_rating.pckl', 'rb')
TF_IDF_predict_rating_result = pickle.load(f)
f.close()
```

```python
test['TF-IDF-pred'] = TF_IDF_predict_rating_result    
test
```

```python
se = 0
for index, row in test.iterrows():
    e = row.rating - row['TF-IDF-pred']
    se += e*e
np.sqrt(se/test.shape[0])    
```

```python
combine = []

for index, row in test.iterrows():
    sum = 0.6*row['KNN-pred'] + 0.4*row['TF-IDF-pred']
    combine.append(sum)
```

```python
test['combine'] = combine
```

```python
test
```

```python
se = 0
for index, row in test.iterrows():
    e = row.rating - row['combine']
    se += e*e
np.sqrt(se/test.shape[0])    
```
