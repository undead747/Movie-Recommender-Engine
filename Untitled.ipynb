```python
import pandas as pd
import ast
import numpy as np
```

```python
smd = pd.read_csv('../the-movies-dataset/movies_metadata_equal_ratings.csv')
```

```python
smd.info()
```

```python
links_small = pd.read_csv('../the-movies-dataset/links_small.csv')
links_small = links_small[links_small['tmdbId'].notnull()]['tmdbId'].astype('int')
```

```python
smd = smd[smd['id'].isin(links_small)]
```

```python
smd
```

```python
credits = pd.read_csv('../the-movies-dataset/credits.csv')
```

```python
credits1 = credits[credits['id'].isin(links_small)]
```

```python
credits1
```

```python
credits1 = credits1.drop(credits1[credits1.duplicated(['id'])].index)
```

```python
credits1
```

```python
smd = smd.merge(credits1, on='id')
```

```python
smd
```

```python
smd['cast'] = smd['cast'].apply(ast.literal_eval)
```

```python
smd['crew'] = smd['crew'].apply(ast.literal_eval)
```

```python
smd['cast'][0]
```

```python
smd['castname'] = smd['cast'].apply(lambda x: [i['name'] for i in x] if isinstance(x, list) else [])
smd['castimage'] = smd['cast'].apply(lambda x: [i['profile_path'] for i in x] if isinstance(x, list) else [])
smd['character'] = smd['cast'].apply(lambda x: [i['character'] for i in x] if isinstance(x, list) else [])
```

```python
smd['character']
```

```python
cast = pd.DataFrame() 
```

```python
cast['castname'] = smd.apply(lambda x: pd.Series(x['castname']),axis=1).stack().reset_index(level=1, drop=True)
```

```python
cast
```

```python
cast['character'] = smd.apply(lambda x: pd.Series(x['character']),axis=1).stack().reset_index(level=1, drop=True)
```

```python
cast
```

```python
def fill_na(x):
    result = []
    for i in x:
        if i is None:
            i = '/https://icon-library.net/images/default-user-icon/default-user-icon-8.jpg'
        result.append(i)    
    return result               
```

```python
fill_na(smd['castimage'][0])
```

```python
smd['castimage'] = smd['castimage'].apply(fill_na)
```

```python
smd['castimage'][0]
```

```python
cast['image'] = smd.apply(lambda x: pd.Series(x['castimage']),axis=1).stack().reset_index(level=1, drop=True)
```

```python
cast['index'] = cast.index
```

```python
cast = cast[['index','castname','character','image']]
```

```python
cast = cast.reset_index(drop=True)
```

```python
cast['id'] =cast.index
```

```python
cast = cast[['id','index','castname','character','image']]
```

```python
def get_director(x):
    for i in x:
        if i['job'] == 'Director':
            return i['name']
    return np.nan
```

```python
smd['director'] = smd['crew'].apply(get_director)
```

```python
director = pd.DataFrame()
director['director'] = smd['director']
```

```python
director['index'] = director.index
director = director[['index','director']]
director
```

```python
director.to_csv(r'../for database/director.csv', index=False)
cast.to_csv(r'../for database/cast.csv', index=False)
```
